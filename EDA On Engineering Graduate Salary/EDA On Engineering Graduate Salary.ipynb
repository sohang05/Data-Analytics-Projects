### Now let's check the specialization column

```python
df.Specialization.value_counts()
```

### If we observe this, we can understand that there are a lot of single values which can unnecessarily affect the model if we try to train it using this data.

```python
specialization = df.Specialization.value_counts()
lessthan10 = specialization[specialization<=10]
lessthan10
```

```python
def remove_specialization_lessthan10(value):
    if value in lessthan10:
        return 'Other'
    else:
        return value

df['Specialization'] = df['Specialization'].apply(remove_specialization_lessthan10)

df['Specialization'].value_counts()
```

### Let's add an Age column

```python
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day)<(born.month,born.day))

df['Age'] = df['DOB'].apply(age)

df
```

### Placing the Age column at the beginning instead of keeping it at the end

```python
df.insert(2,'Age',df.pop('Age'))

df
```

```python
df.drop('DOB',axis=1,inplace = True)
```

### In the above charts we can definitely see some outliers, so we can take care of it by replacing the unnecessary data.

```python
df = df.replace(-1,np.nan)
```

### Now again check for null values.

```python
df.isnull().sum()
```

### Fill up the null values

```python
cols_with_nan = [column for column in df.columns if df.isna().sum()[column]>0]

for column in cols_with_nan:
    df[column] = df[column].fillna(df[column].mean())
```
